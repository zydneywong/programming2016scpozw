```r
require(MASS)

# Set the correlation parameter and mean
beta = 0.5
SIGMA = matrix(c(1,beta,beta,1),ncol=2)
MU = c(2.0,1.0)

# Set the sample size
N = 50

# Draw your sample
out <- mvrnorm(N, mu = MU, Sigma = SIGMA)
```

```r
# Look at a section of the data
dim(out)
out[1:10,]
```

```r
# Plot the random variables in the x-y plane
plot(out)
```

```r
plot(out)
abline(lm(out[,2]~out[,1]),col="red") # regression line (y~x)
```

```r
# A loop sum function

loop_sum <- function(x){
    sum <- 0
    for(i in 1:length(x)){
        sum <- sum + x[i]
        }
    return(sum)
    }
x <- rep(1, 1e1)

system.time(loop_sum(x))
system.time(sum(x))
system.time(rep(1,length(x))%*%x)
```

```r
x <- rep(1,10000000)

system.time(sum(x))
system.time(x%*%x)
```

```r
# Data set is named 'out', which we split into y and X
y <- out[, 2]
X <- out[, 1]

# Intermediate calculations
XT = t(X)
XTX = XT%*%X
invXTX = solve(XTX)
XTy = XT%*%y
beta = invXTX %*% XTy
beta

# Add line to plot
plot(out)
abline(lm(out[,2]~out[,1]),col="red") # regression line y~x
abline(a=0, b=beta, col='blue')
```

```r
system.time(solve(XTX) %*% XTy)
system.time(solve(XTX, XTy))

beta2 = solve(XTX, XTy)
plot(out)
abline(lm(out[,2]~out[,1]), col="red") # regression line (y~x) 
abline(a=0, b=beta, col="blue")
abline(a=0, b=beta2, col="green")
```
